### CPTSketchGraphs
In this notebook enables visualisation of constraint-preserving transformations included in the **CPTSketchGraphs** dataset.

```python
import sys
sys.path.append('SketchGraphs/sketchgraphs')
from utils import sketches_to_renderings, show_cpts
from sketchgraphs.data import flat_array
import matplotlib.pyplot as plt
```

First, lets load the **CPTSketchGraphs** dataset. Please follow the [instruction](README.md#cptsketchgraphs-dataset) on downloading the data. This notebook assumes the data file is already downloaded and located inside the `CPTSketchGraphs\` directory.

```python
seq_path = "data/cptsketchgraphs.npy"
sequences = flat_array.load_dictionary_flat(seq_path)
sequences

cpts_per_sketch = 50
print('Total Number of CPTs: ' + str(len(sequences['keys']) * cpts_per_sketch))
```

We will now load all availiable cpts for the CAD sketch with id `sample_key`.

```python
sample_key = 1076066
cpts = sequences['perturbations'][sample_key]
print('Total Number of CPTs: ' + str(len(cpts)))
```

We visualize the different cpt augmentations

```python
number_cpts_to_show = 10
imgs = sketches_to_renderings(cpts[:number_cpts_to_show])
show_cpts(imgs, rows = 2, cols = 5)
```

We can also visualize the original CAD sketch from SketchGraphs. This notebook assumes the original SketchGraphs dataset is already downloaded and located inside the `CPTSketchGraphs\` directory as described [here](README.md#cptsketchgraphs-dataset).

```python
seq_path = "data/sg_filtered_unique.npy"
sequences = flat_array.load_dictionary_flat(seq_path)
```

We render the original CAD sketch with id `sample_key`.

```python
original_sketch = sequences['sequences'][sample_key]
image = sketches_to_renderings([original_sketch])
```

```python
plt.figure(figsize=(6, 6))  
plt.imshow(image[0])
plt.axis('off')
plt.show()
```
